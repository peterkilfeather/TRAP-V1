```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")


fasn_coverage = pd.read_csv("fasn.coverage", sep='\t', header=None)
fasn_coverage.columns = ['chr', 'coordinate', 'reads']
fasn_coverage['coordinate'] = fasn_coverage['coordinate'] - fasn_coverage.at[0, 'coordinate']
fasn_coverage.head()
```

```python
sns.relplot(x="coordinate", y="reads", data=fasn_coverage)
```

```python
g = sns.relplot(x="coordinate", y="reads", kind="line", data=fasn_coverage)
g.fig.autofmt_xdate()
```

```python
g = sns.relplot(x="coordinate", y="reads", kind="line", data=fasn_coverage)
g.fig.suptitle('Read coverage spanning Fasn locus')
```
